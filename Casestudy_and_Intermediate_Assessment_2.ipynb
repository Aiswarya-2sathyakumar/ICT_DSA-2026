```python
import pandas as pd
```

#Load SpaceX Launch data and Rocket metadata from API

```python
import requests
#API url
launch_url =" https://api.spacexdata.com/v4/launches"
rocket_url ="https://api.spacexdata.com/v4/rockets"
# Fetches data from API
launch_data = requests.get(launch_url).json()
rocket_data = requests.get(rocket_url).json()
#Converts data into dataframe
launch_df = pd.DataFrame(launch_data)
rocket_df = pd.DataFrame(rocket_data)
```

##Extract relevant columns

```python
launch_df = launch_df[['name', 'date_utc', 'success', 'details', 'rocket']]
print(launch_df)
```

##Convert date_utc to datetime

```python
launch_df['date_utc'] = pd.to_datetime(launch_df['date_utc'])
print(launch_df['date_utc'])
```

##Extract the year

```python
launch_df['year'] = launch_df['date_utc'].dt.year
print(launch_df['year'])
```

##Extract relevant columns

```python
rocket_df = rocket_df[["id", "name", "type", "active", "stages"]]
print(rocket_df)
```

```python
rocket_df.rename(columns={'id': 'rocket'}, inplace=True)
rocket_df.rename(columns={'name': 'rocket_name'}, inplace=True)
print(rocket_df)
```

#Merge Launch and Rocket Data


```python
merge_df = pd.merge(launch_df, rocket_df ,on ='rocket',how = 'left')    #left(keep all launch records)
print(merge_df.head(5))
```

#Add Simulated Country Information

```python
import numpy as np
countries =['USA', 'Russia', 'India', 'China', 'France']
merge_df['country'] = pd.Series(np.random.choice( countries,len(merge_df)))

print(merge_df.head())
```

#Store Merged Data in SQLite3

```python
import sqlite3
conn = sqlite3.connect("spacex.db")

# Store DataFrame as table 'launches'
merge_df.to_sql("launches", conn, if_exists="replace", index=False)

print("Data successfully stored in SQLite database!")

q ="SELECT* FROM launches;"
result = pd.read_sql_query (q,conn)
result
```

##1.Launches by Country


```python
q1 = "SELECT country, COUNT(*) AS total_launches FROM launches GROUP BY country\
 ORDER BY total_launches DESC;"
result = pd.read_sql_query (q1,conn)
result
```

##2. Which year had the highest number of launches?

```python
q2 = "SELECT year,COUNT(*) AS total_launches FROM launches GROUP BY year \
 ORDER BY total_launches DESC LIMIT 1;"
result = pd.read_sql_query (q2,conn)
result
```

##3. Top 5 Missions by Launch Count

```python
q3 = "SELECT name, COUNT(*) AS launch_count FROM launches GROUP BY name ORDER BY launch_count DESC LIMIT 5;"
result = pd.read_sql_query (q3,conn)
result
```
